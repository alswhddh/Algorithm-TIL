```python
def solution(L, x):
    
    #init
    lower = 0
    upper = len(L) - 1
    
    while(lower <= upper):
        middle = int((lower + upper) / 2)
        if L[middle] == x:
            index = middle
            return index
        elif L[middle] < x:
            lower = middle + 1
        else :
            upper = middle - 1

    return -1
```

```python
# 정상 테스트
L1 = [2, 3, 5, 6, 9, 11, 15]
target1 = 6

# 만약 요소가 없을 경우
L2 = [2, 5, 7, 9, 11]
target2 = 4
```

```python
print(solution(L1 , target1))
print(solution(L2, target2))
```
